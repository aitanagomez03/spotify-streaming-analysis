```python
import pandas as pd
import plotly.express as px

# Use a clean white theme for all charts
px.defaults.template = "plotly_white"

print("Ready to go!")
```

```python

# Load both JSON files and stack them together
df = pd.concat([
    pd.read_json("StreamingHistory_music_0.json"),
    pd.read_json("StreamingHistory_music_1.json"),
], ignore_index=True)

# Convert endTime from text to a datetime object
df["endTime"] = pd.to_datetime(df["endTime"])

# Milliseconds → minutes (easier to understand)
df["minutesPlayed"] = df["msPlayed"] / 60_000

print(f"Loaded {len(df):,} rows")
df.head(10)
```
